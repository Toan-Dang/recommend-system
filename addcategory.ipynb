```python
import pandas as pd
```

```python
df = pd.read_csv("Products.csv")
```

```python
suffix = 1
for i in range(len(df)):
    if i % 984 == 0:
        df[i:i+984].to_csv(f"product1_{suffix}.csv", sep =',', index=False, index_label=False)
        suffix += 1
```

```python
df1 = pd.read_csv('product1_1.csv')
df1["Type"] = "Điện thoại"
df1["CategoryName"]= df1['ProductName'].str.split().str[0]
df1.to_csv("product2_1.csv", index=False)

df2 = pd.read_csv('product1_2.csv')
df2["Type"] = "Laptop"
df2["CategoryName"] = df2['ProductName'].str.split().str[0]
df2.to_csv("product2_2.csv", index=False)
```

```python
df1 = pd.read_csv('product2_1.csv')
df2 = pd.read_csv('product2_2.csv')
df = pd.concat([df1,df2], ignore_index= True)
del df1,df2
df.to_csv('Products2.csv',index= False)
#df.to_json('Products2.json', orient='records')
df
```
